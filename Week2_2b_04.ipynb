**PERCOBAAN 3 Menampilkan citra Grayscale**

```python
img_gray = cv2.imread("/content/drive/MyDrive/Colab Notebooks/PCVK25_3F_04/KTM.jpg", cv2.IMREAD_GRAYSCALE)

# Tanpa cmap
plt.subplot(1,2,1)
plt.imshow(img_gray)
plt.title("Tanpa cmap")

# Dengan cmap
plt.subplot(1,2,2)
plt.imshow(img_gray, cmap="gray")
plt.title("Dengan cmap")

plt.show()
```

```python
plt.imshow(img_gray, cmap="magma")
```
